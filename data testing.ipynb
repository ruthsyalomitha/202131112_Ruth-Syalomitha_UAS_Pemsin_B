```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Memuat data testing
test_data = pd.read_csv('heart.csv') 
x_test = test_data[['age', 'thalach']]
```

```python
x_ = test_data['age']
y_ = test_data['thalach']
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#praproses
scaler = StandardScaler()
x_test = scaler.fit_transform(x_test)
x_test[:5]
```

```python
# Melakukan scaling pada data testing
x_test_scaled = scaler.transform(x_test)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_test_scaled)
```

```python
kmeans_test_labels = kmeans.predict(x_test_scaled)
```

```python
kmeans_test_labels
```

```python
# Visualisasi hasil clustering pada data testing

label = kmeans_test_labels
plt.scatter(x_, y_, c= label, cmap='plasma')
plt.title('Clustering pada Data Testing')
plt.xlabel('Age')
plt.ylabel('thalach')
plt.show()

```
